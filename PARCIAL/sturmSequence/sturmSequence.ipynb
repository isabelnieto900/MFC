```python
import matplotlib.pyplot as plt
import numpy as np

# Cargamos los datos.
# unpack=True nos da las columnas por separado
# El comando de Fortran escribe espacios en blanco,
# la mayoría de librerías de Python los ignoran o los usan para separar 'segments'
data = np.loadtxt('LITIO_STURM0.DAT')
r = data[:, 0]
u = data[:, 1]

# Como hay varios bloques, una forma fácil de graficar
# es detectar donde termina el radio y vuelve a empezar
indices = np.where(np.diff(r) < 0)[0]
starts = [0] + (indices + 1).tolist()
ends = indices.tolist() + [len(r)]

plt.figure(figsize=(10, 6))
for i, (start, end) in enumerate(zip(starts, ends)):
    plt.plot(r[start:end], u[start:end], label=f'Estado {i+1}s')

plt.axhline(0, color='black', lw=1)
plt.xlabel('Radio r (u.a.)')
plt.ylabel('u(r) = r * R(r)')
plt.title('Funciones de Onda del Átomo de Litio')
plt.xlim(0, 25) # Ajusta según veas la curva
plt.legend()
plt.grid(True)
plt.show()
```

## Python

```python
import numpy as np

def contar_sturm(diag, subdiag, e_test):
    """Cuenta cuántos autovalores son menores que e_test."""
    count = 0
    n = len(diag)

    # Primer elemento de la secuencia
    q = diag[0] - e_test
    if q < 0:
        count += 1

    # Recurrencia de Sturm: q_i = (d_i - E) - b_{i-1}^2 / q_{i-1}
    for i in range(1, n):
        if abs(q) < 1e-30:
            q = 1e-30
        q = (diag[i] - e_test) - (subdiag[i-1]**2) / q
        if q < 0:
            count += 1
    return count

def buscar_energia(diag, subdiag, n_estado, e_min, e_max, tol=1e-12):
    """Bisección para encontrar el nivel n_estado."""
    low, high = e_min, e_max
    while (high - low) > tol:
        mid = 0.5 * (low + high)
        if contar_sturm(diag, subdiag, mid) >= n_estado:
            high = mid
        else:
            low = mid
    return 0.5 * (low + high)

def solver_tridag(sub, diag, sup, r):
    """Algoritmo de Thomas para la función de onda (Iteración Inversa)."""
    n = len(r)
    u = np.zeros(n)
    gam = np.zeros(n)

    bet = diag[0]
    u[0] = r[0] / bet
    for j in range(1, n):
        gam[j] = sup[j-1] / bet
        bet = diag[j] - sub[j-1] * gam[j]
        u[j] = (r[j] - sub[j-1] * u[j-1]) / bet
    for j in range(n-2, -1, -1):
        u[j] -= gam[j+1] * u[j+1]

    # Retorna vector normalizado
    return u / np.sqrt(np.sum(u**2))
```

```python
import time
# 1. Parámetros del sistema
npt = 6000
d_max = 25.0
h = d_max / npt
alfa = 2.535930  # El valor de Fortran
l_angular = float(input("MOMENTUM ANGULAR (l) = "))
start_time = time.perf_counter()
# Coordenada radial (evitando r=0)
r_coords = np.linspace(h, d_max, npt)

# 2. Construcción del Hamiltoniano Tridiagonal
# H_ii = 1/h^2 + V(r)
v_c = -(1.0/r_coords) - (2.0/r_coords)*(1.0 + alfa*r_coords)*np.exp(-2.0*alfa*r_coords)
v_l = l_angular * (l_angular + 1.0) / (2.0 * r_coords**2)
pot_total = v_c + v_l

diag = (1.0 / h**2) + pot_total
subdiag = np.full(npt - 1, -1.0 / (2.0 * h**2))
supdiag = np.copy(subdiag)

# 3. Cálculo y exportación
print(f"\nCalculando con Sturm... (h = {h:.6f})")
print("-" * 45)

with open('LITIO_STURM_PY.DAT', 'w') as f:
    for n in range(1, 5):
        # Búsqueda de energía
        e_au = buscar_energia(diag, subdiag, n, -50.0, 0.0)
        e_ev = e_au * 27.211

        print(f"N={n} | E(au)={e_au:12.8f} | E(eV)={e_ev:12.8f}")

        # Obtener función de onda
        vk = np.ones(npt)
        diag_shift = diag - e_au
        for _ in range(15):
            vk = solver_tridag(subdiag, diag_shift, supdiag, vk)

        # Normalización Física
        # Pasamos de suma(vk^2)=1 a integral(psi^2)=1
        norma_v = np.sqrt(np.sum(vk**2))
        vk = (vk / norma_v) / np.sqrt(h)

        # Corrección de Fase (Para que no salga invertida)
        # Identificamos el signo del primer lóbulo significativo
        for val in vk:
            if abs(val) > 1e-5:
                if val < 0.0:
                    vk = -vk
                break
        # Escribir al .DAT (r, psi)
        for i in range(0, npt, 10):
            f.write(f"{r_coords[i]:12.8f} {vk[i]:12.8f}\n")
        f.write("\n") # Separador de bloques para estados
end_time = time.perf_counter()   # Fin
print(f"Tiempo de ejecución: {end_time - start_time:.6f} segundos")
print("-" * 45)
print("¡Listo! Archivo 'LITIO_STURM_PY.DAT' generado.")
```
