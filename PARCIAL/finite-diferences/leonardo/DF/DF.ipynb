```python
import numpy as np
import math

def mnbrack(ax, bx, pcom, xicom, func):
    """Busca un intervalo que encierre un mínimo local de la función."""
    gold = 1.618034
    tiny = 1.0e-20

    fa = func(pcom + ax * xicom)
    fb = func(pcom + bx * xicom)

    if fb > fa:
        ax, bx = bx, ax
        fa, fb = fb, fa

    cx = bx + gold * (bx - ax)
    fc = func(pcom + cx * xicom)

    while fb > fc:
        r = (bx - ax) * (fb - fc)
        q = (bx - cx) * (fb - fa)

        # Denominador seguro para evitar división por cero
        denom = 2.0 * math.copysign(max(abs(q - r), tiny), q - r)
        u = bx - ((bx - cx) * q - (bx - ax) * r) / denom
        fu = func(pcom + u * xicom)

        if (bx - u) * (u - cx) > 0.0:
            if fu < fc:
                ax = bx; bx = u; fa = fb; fb = fu
                return ax, bx, cx, fa, fb, fc
            elif fu > fb:
                cx = u; fc = fu
                return ax, bx, cx, fa, fb, fc

        ax = bx; bx = cx; cx = cx + gold * (cx - bx)
        fa = fb; fb = fc; fc = func(pcom + cx * xicom)

    return ax, bx, cx, fa, fb, fc

def brent_opt(ax, bx, cx, pcom, xicom, func):
    """Encuentra el mínimo exacto usando el método de Brent en 1D."""
    itmax = 100
    cgld = 0.3819660
    eps = 3.0e-8

    a = min(ax, cx)
    b = max(ax, cx)
    x = bx; w = bx; v = bx
    fw = func(pcom + x * xicom); fv = fw; fx = fw
    e = 0.0
    d = 0.0

    for _ in range(itmax):
        xm = 0.5 * (a + b)
        tol1 = eps * abs(x) + 1.0e-10
        tol2 = 2.0 * tol1

        if abs(x - xm) <= (tol2 - 0.5 * (b - a)):
            break

        if abs(e) > tol1:
            r = (x - w) * (fx - fv)
            q = (x - v) * (fx - fw)
            p = (x - v) * q - (x - w) * r
            q = 2.0 * (q - r)
            if q > 0.0: p = -p
            q = abs(q)
            etemp = e
            e = d

            if abs(p) >= abs(0.5 * q * etemp) or p <= q * (a - x) or p >= q * (b - x):
                e = (b - x) if x < xm else (a - x)
                d = cgld * e
            else:
                d = p / q
                u = x + d
                if (u - a) < tol2 or (b - u) < tol2:
                    d = math.copysign(tol1, xm - x)
        else:
            e = (b - x) if x < xm else (a - x)
            d = cgld * e

        u = x + d if abs(d) >= tol1 else x + math.copysign(tol1, d)
        fu = func(pcom + u * xicom)

        if fu <= fx:
            if u >= x: a = x
            else: b = x
            v = w; fv = fw; w = x; fw = fx; x = u; fx = fu
        else:
            if u < x: a = u
            else: b = u
            if fu <= fw or w == x:
                v = w; fv = fw; w = u; fw = fu
            elif fu <= fv or v == x or v == w:
                v = u; fv = fu

    return x, fx

def linmin(p, xi, func):
    """Minimización a lo largo de una línea (dirección dada por xi)."""
    ax = 0.0
    xx = 1.0
    ax, xx, bx, fa, fx, fb = mnbrack(ax, xx, p, xi, func)
    xmin, fret = brent_opt(ax, xx, bx, p, xi, func)

    xi_new = xi * xmin
    p_new = p + xi_new
    return p_new, xi_new, fret

def powell(p, xi, ftol, func):
    """Método de Powell para minimizar una función multivariable."""
    n = len(p)
    pt = np.copy(p)
    fret = func(p)
    iteraciones = 0
    itmax = 200

    while True:
        iteraciones += 1
        fp = fret
        ibig = 0
        del_val = 0.0

        for i in range(n):
            xit = np.copy(xi[:, i])
            fptt = fret
            p, xit, fret = linmin(p, xit, func)

            if abs(fptt - fret) > del_val:
                del_val = abs(fptt - fret)
                ibig = i

        if 2.0 * abs(fp - fret) <= ftol * (abs(fp) + abs(fret)) + 1.0e-20:
            return p, xi, iteraciones, fret

        if iteraciones == itmax:
            print("Advertencia: Powell excedió el máximo de iteraciones.")
            return p, xi, iteraciones, fret

        ptt = 2.0 * p - pt
        xit = p - pt
        fptt = func(ptt)

        if fptt < fp:
            t = 2.0 * (fp - 2.0 * fret + fptt) * (fp - fret - del_val)**2
            if t < del_val * (fp - fptt)**2:
                p, xit, fret = linmin(p, xit, func)
                xi[:, ibig] = np.copy(xi[:, n-1])
                xi[:, n-1] = xit

        pt = np.copy(p)
```

```python
import numpy as np

def zbrak(fx, x1, x2, n):
    """Busca intervalos donde la función cambia de signo (cruza el cero)."""
    xb1 = []
    xb2 = []
    dx = (x2 - x1) / n
    x = x1
    fp = fx(x)

    for _ in range(n):
        x += dx
        fc = fx(x)
        if fc * fp < 0.0:
            xb1.append(x - dx)
            xb2.append(x)
        fp = fc

    return xb1, xb2

def zbrent(fx, x1, x2, tol):
    """Encuentra la raíz exacta usando el método de Brent."""
    itmax = 100
    eps = 3.0e-8

    a = x1; b = x2; c = x2
    fa = fx(a); fb = fx(b)

    if (fa > 0.0 and fb > 0.0) or (fa < 0.0 and fb < 0.0):
        raise ValueError("La raíz no está acotada en el intervalo dado para zbrent.")

    fc = fb
    for _ in range(itmax):
        if (fb > 0.0 and fc > 0.0) or (fb < 0.0 and fc < 0.0):
            c = a; fc = fa
            e = b - a; d = b - a

        if abs(fc) < abs(fb):
            a = b; b = c; c = a
            fa = fb; fb = fc; fc = fa

        tol1 = 2.0 * eps * abs(b) + 0.5 * tol
        xm = 0.5 * (c - b)

        if abs(xm) <= tol1 or fb == 0.0:
            return b

        if abs(e) >= tol1 and abs(fa) > abs(fb):
            s = fb / fa
            if a == c:
                p = 2.0 * xm * s
                q = 1.0 - s
            else:
                q = fa / fc; r = fb / fc
                p = s * (2.0 * xm * q * (q - r) - (b - a) * (r - 1.0))
                q = (q - 1.0) * (r - 1.0) * (s - 1.0)
            if p > 0.0: q = -q
            p = abs(p)

            min1 = 3.0 * xm * q - abs(tol1 * q)
            min2 = abs(e * q)

            if 2.0 * p < min(min1, min2):
                e = d; d = p / q
            else:
                d = xm; e = d
        else:
            d = xm; e = d

        a = b; fa = fb
        if abs(d) > tol1:
            b += d
        else:
            b += np.copysign(tol1, xm)
        fb = fx(b)

    return b

def tridag(a, b, c, r_vec):
    """Resuelve un sistema de ecuaciones tridiagonal."""
    n = len(r_vec)
    u = np.zeros(n)
    gam = np.zeros(n)

    if b[0] == 0.0:
        raise ValueError("Error en tridag: b[0] es 0.")

    bet = b[0]
    u[0] = r_vec[0] / bet

    for j in range(1, n):
        gam[j] = c[j-1] / bet
        bet = b[j] - a[j] * gam[j]
        if bet == 0.0:
            raise ValueError("Error en tridag: bet es 0.")
        u[j] = (r_vec[j] - a[j] * u[j-1]) / bet

    for j in range(n-2, -1, -1):
        u[j] -= gam[j+1] * u[j+1]

    return u

def norma_fisica(v, h):
    """Normaliza la función de onda asegurando que la integral |u(r)|^2 dr = 1"""
    suma = np.sum(v**2)
    factor = np.sqrt(suma * h)
    return v / factor
```

```python
import numpy as np
import time

start_time = time.perf_counter()
# Asumimos que las funciones de optimizacion.py ya están cargadas en memoria

def calcular_variacional_litio():
    # --- 1. CONFIGURACIÓN ---
    try:
        z_global = float(input("CUAL ES EL VALOR DE Z (ej. 3.0 para Litio): "))
    except ValueError:
        print("Entrada no válida. Usando Z = 3.0 por defecto.")
        z_global = 3.0

    # Tolerancia muy fina para resultados de nivel "paper"
    ftol = 1.0e-12

    # --- 2. FUNCIONES OBJETIVO FÍSICAS ---
    def func(x):
        """Energía del estado base (Ecuación 7 del paper)."""
        a = x[0]
        b = x[1]
        y = 2.0 * a * b / ((2.0 * a + b)**5)

        energia = (a**2 - 2.0 * z_global * a + (5.0 / 8.0) * a +
                   (1.0 / 8.0) * b**2 - (z_global * b / 4.0) +
                   y * (8.0 * a**4 + 20.0 * a**3 * b + 12.0 * a**2 * b**2 +
                        10.0 * a * b**3 + b**4))
        return energia

    def ip_func(alfa, beta):
        """Energía de Ionización / Electrón de Valencia (Teorema de Koopmans)."""
        a2 = alfa * alfa
        b2 = beta * beta
        y = 2.0 * alfa * beta / ((2.0 * alfa + beta)**5)

        ip = ((1.0 / 8.0) * b2 - (z_global * beta / 4.0) +
              y * (8.0 * a2 * a2 + 20.0 * a2 * alfa * beta + 12.0 * a2 * b2 +
                   10.0 * alfa * beta * b2 + b2 * b2))
        return ip

    # --- 3. OPTIMIZACIÓN ---
    # Inicialización de la matriz de direcciones (Identidad 2x2)
    xi = np.eye(2)

    # Valores iniciales para Alfa y Beta [Ajustados a 2.69 y 0.70 para mayor estabilidad]
    p_inicial = np.array([2.69, 0.70])

    print("Calculando parámetros óptimos con el método de Powell...")
    p_optimo, xi_final, iteraciones, energia_minima = powell(p_inicial, xi, ftol, func)

    alfa_opt = p_optimo[0]
    beta_opt = p_optimo[1]

    # --- 4. RESULTADOS ---
    factor_ev = 27.211386 # Factor CODATA de alta precisión

    print('\n' + '='*40)
    print(f' ITERACIONES:           = {iteraciones}')
    print(f' ALFA, BETA             = {alfa_opt:.6f}, {beta_opt:.6f}')
    print('-'*40)
    print(f' ENER. ESTADO BASE (au) = {energia_minima:.6f}')
    print(f' ENER. ESTADO BASE (eV) = {factor_ev * energia_minima:.6f}')
    print('-'*40)
    print(f' ENER. IONIZACIÓN (au)  = {ip_func(alfa_opt, beta_opt):.6f}')
    print(f' ENER. IONIZACIÓN (eV)  = {factor_ev * ip_func(alfa_opt, beta_opt):.6f}')
    print('='*40)

# Ejecutar el cálculo
calcular_variacional_litio()
end_time = time.perf_counter()   # Fin
print(f"Tiempo de ejecución: {end_time - start_time:.6f} segundos")
```

```python
import numpy as np
import time

start_time = time.perf_counter()
# --- 1. DEFINICIÓN DE LA FÍSICA (Determinante y Potencial) ---

def determ(x_energia, h, ele, npt):
    alfa = 2.535930  # <-- VUELVE AL ALFA DE C++
    h2 = h * h
    p0 = 0.0
    p1 = 1.0

    for i in range(1, npt + 1): # <-- HASTA npt (INCLUSIVE)
        r = h * i
        pot_val = -(1.0/r) - (2.0/r)*(1.0 + alfa*r)*np.exp(-2.0*alfa*r) + \
                  ele*(ele + 1.0)/(2.0*r**2)

        if i == 1: # <-- VUELVE EL IF QUE TIENE C++
            p1 = (2.0 + 2.0 * h2 * pot_val - 2.0 * x_energia * h2)
        else:
            p2 = (2.0 + 2.0 * h2 * pot_val - 2.0 * h2 * x_energia) * p1 - p0
            p0 = p1
            p1 = p2

        if abs(p1) > 1e20:
            p1 *= 1e-20
            p0 *= 1e-20

    return p1

def pot_local(x, l):
    alfa = 2.535930 # <-- VUELVE AL ALFA DE C++
    pot = -(1.0/x) - (2.0/x) * (1.0 + alfa*x) * np.exp(-2.0*alfa*x)
    pot += l * (l + 1.0) / (2.0 * x**2)
    return pot

# --- 2. RUTINA PRINCIPAL ---

def calcular_espectro():
    # Parámetros de alta precisión
    npt = 6000
    d = 25.0
    x1 = -10.0
    x2 = 0.0
    n_divs = 20000
    tol = 1.0e-12

    ele = float(input("MOMENTUM ANGULAR (l) = "))
    h = d / npt

    determ_wrapper = lambda x: determ(x, h, ele, npt)

    print("Buscando intervalos...")
    # Asumo que zbrak y zbrent ya están definidos en tu entorno o importados
    xb1, xb2 = zbrak(determ_wrapper, x1, x2, n_divs)

    num_raices = len(xb1)
    print(f"Se encontraron {num_raices} autovalores potenciales.\n")

    energias = []

    with open("LITIO_PYTHON.DAT", "w") as f:
        for k in range(num_raices):
            print(f"Refinando raíz {k+1}...")
            raiz = zbrent(determ_wrapper, xb1[k], xb2[k], tol)
            energias.append(raiz)

            # --- Construcción de la Matriz Tridiagonal ---
            diag = np.zeros(npt)
            subd = np.zeros(npt)
            superd = np.zeros(npt)
            vk = np.ones(npt)

            h2 = h * h
            for i in range(npt):
                r = h * (i + 1)
                vp = pot_local(r, ele)
                diag[i] = 2.0 + 2.0 * h2 * vp - 2.0 * h2 * raiz
                subd[i] = -1.0
                superd[i] = -1.0

            subd[0] = 0.0
            superd[-1] = 0.0

            # Iteración inversa para encontrar el eigenvector
            for _ in range(20):
                vkm1 = vk.copy()
                vk = tridag(subd, diag, superd, vkm1)
                vk = norma_fisica(vk, h)

            norma_vector = np.linalg.norm(vk)
            vk = (vk / norma_vector) / np.sqrt(h)

            # 2. Corrección de Fase (Reflexión)
            # Buscamos el primer punto que salga del ruido para decidir el signo
            indice_fase = np.where(np.abs(vk) > 1e-5)[0][0]
            if vk[indice_fase] < 0.0:
                vk = -vk
            # --------------------------------------

            # Guardar en archivo
            for i in range(0, npt, 10):
                r_val = float(i+1) * h
                f.write(f"{r_val:.8f}  {vk[i]:.8f}\n")
            f.write("\n")

            print(f"Estado N={k+1} -> E(au) = {raiz:.6f} | E(eV) = {raiz * 27.211:.6f}")

    print("\n¡Cálculo finalizado! Datos guardados en LITIO_PYTHON.DAT")

# Ejecutar el programa
calcular_espectro()
end_time = time.perf_counter()   # Fin
print(f"Tiempo de ejecución: {end_time - start_time:.6f} segundos")
```
